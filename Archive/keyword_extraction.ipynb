## using konlpy to extract keyword for 중단원

```python
# tokenizing sentences into nouns, verbs, etc
from konlpy.tag import Mecab
def extract_keyword(corpus: str, stopwords:list) -> list:
    mecab_tokenizer = Mecab()

    mecab_pos = ["NNG", "NNP", "VV", "VA", "MAG"]

    mecab_keywords = [word for word, tag in mecab_tokenizer.pos(corpus, flatten=True) if len(word) > 1 and tag in mecab_pos and word not in stopwords]
    return mecab_keywords
```

```python
list_stopwords = []
extract_keyword(list_qtid_name[0], list_stopwords)
```

```python
for name in list_qtid_name:
    print(name , "-> 키워드:", extract_keyword(name, list_stopwords))
```

## using Okt to extract keyword for 중단원

```python
from konlpy.tag import Okt

def extract_keyword_okt(raw:str, stopwords:list) -> list:
    okt_tokenizer = Okt()
    # position arguments(or position tags): 내가 뽑아내고 싶은 형태소들
    twitter_pos = ["Noun", "Alpha", "Verb", "Number", "Adverb"]
    okt_list = [
        word for word, tag in okt_tokenizer.pos(raw, norm=True, stem=True)
        # norm: normalize 그랰ㅋㅋ -> 그래ㅋㅋ  # stem: stemming 바뀌나->바뀌다
        if len(word) > 1 and tag in twitter_pos and word not in stopwords
    ]
    return okt_list
```

```python
extract_keyword_okt(list_qtid_name[0], list_stopwords)
```

```python
for name in list_qtid_name:
    print(name , "-> 키워드:", extract_keyword_okt(name, list_stopwords))
```

* 세 가지 tokenizer들을 비교해보았을 때, "원" 같은 키워드를 잡는 건 soynlp였다. 
* 전반적으로 soynlp가 명사를 적절하게 추출했다. 다만 "인수정리" 같은 키워드는 잡아내지 못했다.
* mecab은 형태소를 잘게 쪼개는 경향이 강하다."나머지정리" 경우는 mecab의 경우 합성어 분리 ("나머지", "정리")로 했으나, Okt랑 soynlp는 "나머지정리"로 인식했다.
* 따라서 단원 별 키워드 라벨링을 일단 soynlp를 사용할 것이다

```python
# 중단원-소단원 id만 뽑아내기
df_qtid = pd.DataFrame({})
df_qtid["qtid"] = df_quanda["qtid"].sort_values()
df_qtid
```

```python
# 서로 같은 중단원-소단원 id들은 합치기. 총 592개의 소단원들이 남는다.
df_temp = df_qtid.groupby(["qtid"]).agg('sum').reset_index()
df_temp
```

```python
# Split qtid column into 학년, 중단원, 소단원 three independent columns
df_splited_temp = df_temp["qtid"].str.split("-", expand=True,)
df_splited_temp[:5]

df_temp[["학년", "중단원", "소단원"]] = df_splited_temp
df_temp
```

```python
df_qtid = df_temp
df_qtid
```

```python
# 중단원 - [소단원 keyword1, 소단원 keyword2 ...]를 만들기 위해서 학년 - 중단원 column을 따로 만든다
df_temp = df_qtid[["학년", "중단원"]]
df_temp = df_temp.groupby(["학년", "중단원"]).agg('sum').reset_index()
df_temp
```

```python
# [소단원 keyword1, 소단원 keyword2 ...] column value를 list로 만든다.
list_keyword = []
for index, row in df_temp.iterrows():
    print(row)
    arr_qtid_name =df_quanda[(df_quanda["학년"]==row['학년']) & (df_quanda["중단원"]==row['중단원'])]['name'].unique()
    list_qtid_name = list(arr_qtid_name)
    print(list_qtid_name)
    dict_nouns = noun_extractor.train_extract(list_qtid_name, min_noun_frequency=1, min_noun_score=0.3)
    list_nouns = list(dict_nouns.keys())
    list_keyword.append(list_nouns)
```

```python
# [소단원 keyword1, 소단원 keyword2 ...] column value를 살펴보자.
list_keyword
```

```python
# [소단원 keyword1, 소단원 keyword2 ...] value들을 각 중단원 별로 column으로써 추가한다.
df_temp["소단원_keywords_soynlp"] = list_keyword
df_temp
```
